```python
recorder = chemin
```

```python
#recoder l’age en variable qualitative ne faisant des inter
def categories(Age):
    if Age < 18:
        return "Mineur"
    elif Age < 38:
        return "Jeune"
    elif Age < 58:
        return "Adulte"
    else:
        return "Vieux"
    
recorder['Age']=recorder['Age'].map(categories)
recorder
```

```python
somme = chemin.groupby('Country')['Customer spendings'].sum()
somme
```

```python
"""Créer une visualisation des données du graphique à barres qui montre les dépenses
des clients par pays"""

countries = chemin['Country']
#plt.rcParams['figure.figsize'] = (30, 30)
#plt.rcParams['font.size'] = '17'
plt.xticks(rotation = '45')
plt.title("les dépenses des clients par pays")
plt.xlabel('Country')
plt.ylabel('Customer spendings')
#plt.bar(dataset['Country'], dataset['Customer spendings'])
chemin.groupby('Country')['Customer spendings'].sum().plot(kind='bar', x='Country',y='Customer spendings',figsize=(20,10))
#df2.plot(kind='bar')
```

```python
"""Nettoyer les lignes avec moins de 10 € de dépenses par client : supprimez toutes les
lignes d’utilisateurs ayant dépensé moins de 10 € sur le site,
"""
chemin.drop(chemin[chemin["Customer spendings"] < 10].index, inplace=True)
chemin
```

```python
"""Nettoyer les doublons : supprimer toutes les lignes qui apparaissent plus d’une fois
dans les données,"""

chemin.drop_duplicates( keep = 'first' , inplace = True , ignore_index = False )
chemin
```
