```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

"""On considère les trois vecteurs suivants
x = np.linspace(0, 2, 10)
y = x**2
z = x**3"""
#1. Construire une courbe avec x en abscisse et y en ordonnée

```python
x = np.linspace(0, 2, 10)
y = x**2
z = x**3

plt.plot(x, y)
plt.show()
plt.close()
```

"""2. Changer l'épaisseur et la couleur de la ligne du graphique précédemment créé
avec le paramètre "c" et "lw""""

```python
plt.plot(x, y, c="red", lw=4)
plt.show()
plt.close()
```

#3. Construire sur le même graphique deux courbes :

```python
#1. Une première courbe avec x en abscisse et y en ordonnée
#2. Une deuxième courbe avec x en abscisse et z en ordonnée
#3. Ajouter une légende au graphique
#4. Ajouter un titre au graphique
#5. Ajouter un titre à chaque axe


plt.plot(x, y)
plt.plot(x, z)
plt.legend(['courbe 1','courbe 2'], title="legend")
plt.title('les deux courbes')
plt.xlabel('abscsse')
plt.ylabel('ordonnée')
plt.show()
plt.close()
```

```python
#4. Afficher à travers un boxplot la distribution de x
plt.boxplot(x)
plt.show()
plt.close()
```

```python
#5. Afficher à travers un histogramme la distribution de x
plt.hist(x)
plt.show()
plt.close()
```

```python
#6. Construire le nuage de points entre x et y
plt.scatter(x,y)
plt.show()
plt.close()
```

```python
"""10. Affichez le nuage de point entre la variable "ÂGE" et la variable "DÉPRESSION"
puis utiliser la variable ‘COEUR” pour la couleur des points"""
plt.scatter(df["AGE"],df["DEPRESSION"],c=df["CŒUR"])
plt.show()
plt.close()
```

```python
"""11. Stockez dans "liste qual" les noms des variables qualitatives et dans “liste_quant”
celles des variables quantitatives."""
liste_qual = ["SEXE", "TDT", "GAJ", "CŒUR", "ECG", "ANGINE", "PENTE"]
liste_quant = ["AGE", "PAR", "CHOLESTEROL", "FCMAX", "DEPRESSION"]
```

```python
"""12. En utilisant “liste_quant”, faites une représentation graphique de chaque variable
quantitative en utilisant une boucle for."""
for quant in liste_quant:
    plt.hist(df[quant],edgecolor="black",label=quant)
    plt.title(quant)
    plt.legend()
    plt.show()
    plt.close()
```

```python
"""13. En utilisant “liste_qual”, faites une représentation graphique de chaque variable
quatitative en utilisant une boucle for"""
for qual in liste_qual:
    plt.hist(df[qual])
    plt.title(qual)
    plt.show()
    plt.close()
```

```python
plt.figure(figsize=(30, 15))
plt.gcf().subplots_adjust(wspace=1, hspace=1)
i=1
for quant in liste_qual:
    plt.subplot(3,3,i)
    x=df[quant].value_counts()
    x.plot(kind='bar')
    plt.title(quant, fontsize=40)
    #plt.xlabel(fontsize=20)
    plt.xticks(fontsize=20, rotation=45)
    #plt.ylabel(fontsize=20)
    plt.yticks(fontsize=20)

    i+=1
```
